#### 4. Rename any columns with spaces or special characters to not contain any.

```python
import string
# the string library has default strings that contain all letters or numbers
uppercase = string.ascii_uppercase  # this is for uppercase
lowercase = string.ascii_lowercase  # this is for lowercase
allcase=string.ascii_letters  # this is for all characters
```

```python
# Your code here please

# Hint1: use "for","if-else" and column rename function
# Hint2: compare each column name against allcase (see above)


# here is a startup code to help you :) 

for column in cast.columns:
    temp = column
    #all the code below this "for" 
    for letter in column:
        #replace the letter if it is a special characters with no space
        if letter not in allcase:
            column = column.replace(letter,'')
    cast.rename(columns={temp:column}, inplace=True)
    
# finally print cast columns to confirm if your code was OK or not
print(cast.columns)  
```
